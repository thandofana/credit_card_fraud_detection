# Data Preparation Phase

```python
# separiting the data into two dataframes, one for fraud and the other for legit trasactions

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]

fraud.head()
```

```python
fraud.shape
```

```python
fraud.count()
```

```python
# Using under_sampling to deal with class imbalance, 
# making the number of legit trasaction to be equal to
#fraud trasactions

legit_sample = legit.sample(n =492)
```

```python
# creating a new data frame combing the equalled legit and fraud transactions datasets
# this is the dataset that will  be used when modeling 

new_data = pd.concat([legit_sample,fraud])
new_data.head()
```

```python
# the new dataframe shows thers now equal numbers for fraud and legit
new_data['Class'].value_counts()
```

```python
# defining X and Y (X For the dependant variables and Y for Class)

X = new_data.drop(columns = ['Class'])
y = new_data['Class']
```

# Modelling Phase

```python
# importing the train_test_split function from scikit-learn, it will help
#split the data into four part. X train X test Y train and Y test

from sklearn.model_selection import train_test_split

Xtrain,Xtest,ytrain,ytest = train_test_split(X,y, test_size =0.2)

print(Xtrain.shape, Xtest.shape, ytrain.shape,ytest.shape)
```

```python
# Data modeling using Logistic Regression Algorithm

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(Xtrain,ytrain)
```

# Evaluation Phase

```python
# predicting the outcome/ ytest values by pluggin in xtest data into the model


ypred = model.predict(Xtest)
ypred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,ypred)
```

### the confusion matrix,  shows that from a total of 492 transactions, 93 legit were predicted to be legit . 88 fraud transactions  were correctly predicted to be  fraud. 14 frauds were predicted to be legit and 2 legit transactions were predicted to be fraud.






```python
# calculationg the accuracy of the model and at 92% its 
# exceptional

accuracy = model.score(Xtest, ytest)
accuracy
```
